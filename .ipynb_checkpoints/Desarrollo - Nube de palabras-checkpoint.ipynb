#### Crear Nube de Palabra

```python
#Archivo de texto debe estar en la misma carpeta del archivo
#Icono descargar de fontawesome : https://fontawesome.com/icons
stylecloud.gen_stylecloud(file_path='textobase.txt', 
                          icon_name= 'fas fa-cloud',
                          #colors = '#FFC300',
                          palette = 'colorbrewer.diverging.RdYlBu_7',
                          background_color = 'black',
                          output_name = 'nube.png',
                          collocations = False,
                          custom_stopwords = palabras_irrelevantes)
                                                  

#Automaticamente se genera una imagen en la carpeta
```
